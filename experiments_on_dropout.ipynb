```python
def scale(X, decimals):
    return np.round((X - np.mean(X, axis=0))/ np.std(X, axis=0), decimals=decimals)

seed = 10

np.random.seed(seed)

data = df.values
np.random.shuffle(data)

X = data[:, :-2]
y_temp = data[:, -1].astype(int)

X = scale(X, 4)

y = np.zeros((y_temp.shape[0], 3))
y[np.arange(y_temp.shape[0]), y_temp - 1] = 1

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

X_train, X_test, y_train, y_test = np.array(X_train), np.array(X_test), np.array(y_train), np.array(y_test)
```

```python
#a)

lr = 1e-2
decay = 1e-6
batch_size = 32
epochs = 10000
epochs_interval =100
early_stop_threshold = 1e-2

model = Sequential([
    Dense(10, input_dim=X_train.shape[1], activation='relu', kernel_initializer='lecun_normal', kernel_regularizer=l2(decay)),
    Dense(10, activation='relu', kernel_initializer='he_normal', kernel_regularizer=l2(decay)),
    Dense(3, activation='softmax' ,kernel_initializer='he_normal', kernel_regularizer=l2(decay))
    ])

model.compile(optimizer=optimizers.SGD(lr=lr),
                loss='categorical_crossentropy',
                metrics=['accuracy'])

h = model.fit(X_train, y_train, batch_size=batch_size, epochs=epochs, verbose=1, validation_data=(X_test, y_test), shuffle=True)

# h

        

plt.figure()
plt.plot(h.history['acc'])
plt.plot(h.history['val_acc'])
plt.title('accurazy')
plt.xlabel('epoch')
plt.legend(['train_acc', 'val_acc',], loc='upper left')

```
